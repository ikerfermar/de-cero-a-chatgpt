En los bloques importantes de este cuaderno, sigue esta secuencia:

1. **Texto → tokens** (discretización).
2. **Tokens → embeddings** (representación densa).
3. **Embeddings → modelo secuencial** (RNN/LSTM/GRU).
4. **Modelo → probabilidad del siguiente token**.

Lenguaje autoregresivo:
$$
P(w_{1:T})=\prod_{t=1}^{T}P(w_t\mid w_{<t})
$$

## Guía de lectura (teoría ↔ práctica)

Este cuaderno une 3 bloques de teoría desde cero:

1. **Tokenización**: convertir texto en ids.
2. **Modelos secuenciales**: RNN, LSTM y GRU.
3. **Language Modeling**: predecir el siguiente token.

Objetivo de un LM autoregresivo:
$$
P(w_{1:T}) = \prod_{t=1}^{T} P(w_t\mid w_{<t})
$$

Pérdida típica (NLL):
$$
\mathcal{L} = -\sum_t \log P(w_t\mid w_{<t})
$$

Perplexity:
$$
\mathrm{PPL}=\exp\left(\frac{1}{N}\sum_t -\log P(w_t\mid w_{<t})\right)
$$

**Qué observar**: cómo cambia la PPL y la estabilidad del entrenamiento entre RNN/LSTM/GRU.

## 1) Motivación: texto discreto vs redes neuronales
Un texto es una secuencia de símbolos discretos (caracteres/palabras). Las redes trabajan con **tensores numéricos**.

Formalmente:
$$
x \xrightarrow{\mathcal{T}} (t_1,\dots,t_T) \xrightarrow{\mathcal{E}} (e_1,\dots,e_T),\quad e_i\in\mathbb{R}^d.
$$

- $\mathcal{T}$: tokenizador (determinista) que mapea texto → tokens/IDs
- $\mathcal{E}$: embeddings (matriz entrenable) que mapea IDs → vectores

## 2) Tokenizador más simple: whitespace + vocab
Vamos a construir un tokenizador **de palabras** (whitespace).

### Conceptos clave
- **Vocabulario** $\mathcal{V}$: conjunto finito de tokens
- **IDs**: cada token tiene un entero
- **OOV**: palabras fuera del vocabulario → `<unk>` (si existe)

````python
from collections import Counter

@dataclass
class Vocab:
    # stoi: string -> integer (token a id)
    stoi: Dict[str, int]
    # itos: integer -> string (id a token)
    itos: List[str]

    # Tokens especiales que usaremos en todo el notebook
    pad: str = '<pad>'   # relleno para igualar longitudes
    unk: str = '<unk>'   # palabra desconocida (fuera de vocabulario)
    bos: str = '<bos>'   # begin-of-sequence (inicio)
    eos: str = '<eos>'   # end-of-sequence (fin)

    # Estas propiedades devuelven el id numérico de cada token especial
    @property
    def pad_id(self): return self.stoi[self.pad]
    @property
    def unk_id(self): return self.stoi[self.unk]
    @property
    def bos_id(self): return self.stoi[self.bos]
    @property
    def eos_id(self): return self.stoi[self.eos]

def build_vocab(texts: List[str], min_freq: int = 1) -> Vocab:
    # Counter contará cuántas veces aparece cada palabra en el corpus
    counter = Counter()

    # Recorremos cada texto y sumamos frecuencias palabra a palabra
    for t in texts:
        counter.update(t.strip().split())

    # Tokens especiales al principio para que tengan ids fijos y fáciles de recordar
    specials = ['<pad>', '<unk>', '<bos>', '<eos>']

    # itos = especiales + palabras frecuentes del corpus
    # c >= min_freq filtra palabras muy raras
    itos = specials + [w for w, c in counter.items() if c >= min_freq and w not in specials]

    # stoi es el diccionario inverso: token -> id
    stoi = {w: i for i, w in enumerate(itos)}

    # Devolvemos un objeto Vocab con ambos mapas
    return Vocab(stoi=stoi, itos=itos)

def encode(vocab: Vocab, text: str, add_bos_eos: bool = False) -> List[int]:
    # Separación simple por espacios (tokenización whitespace)
    toks = text.strip().split()

    # Cada token se convierte a su id; si no existe usamos unk_id
    ids = [vocab.stoi.get(w, vocab.unk_id) for w in toks]

    # Opcional: añadir marcadores de inicio y fin
    if add_bos_eos:
        ids = [vocab.bos_id] + ids + [vocab.eos_id]

    return ids

def decode(vocab: Vocab, ids: List[int]) -> str:
    # Convertimos ids de vuelta a texto, útil para depurar
    # Si un id está fuera de rango, mostramos <oov>
    return ' '.join(vocab.itos[i] if 0 <= i < len(vocab.itos) else '<oov>' for i in ids)

def pad_batch(seqs: List[List[int]], pad_id: int) -> Tuple[torch.Tensor, torch.Tensor]:
    # lengths guarda la longitud real de cada secuencia antes del padding
    lengths = torch.tensor([len(s) for s in seqs], dtype=torch.long)

    # T será la longitud máxima dentro del batch
    T = int(lengths.max().item())

    # Creamos una matriz (batch_size, T) rellena con pad_id
    out = torch.full((len(seqs), T), pad_id, dtype=torch.long)

    # Copiamos cada secuencia en su fila (alineada a la izquierda)
    for i, s in enumerate(seqs):
        out[i, :len(s)] = torch.tensor(s, dtype=torch.long)

    # Devolvemos: tensor padded + longitudes originales
    return out, lengths

print('Tokenizer utils ready')
````

## 3) Por qué el padding es necesario
En un batch, las secuencias tienen longitudes distintas. Para formar un tensor rectangular `(B, T)` se usa `<pad>`.

**Error típico**:
```python
torch.tensor([[1,2,3],[1,2]])  # falla
```
porque las filas no tienen la misma longitud.

````python
seqs = [encode(vocab, s, add_bos_eos=True) for s in corpus]
print('Longitudes:', [len(s) for s in seqs])

x_pad, lengths = pad_batch(seqs, vocab.pad_id)
print('x_pad shape:', x_pad.shape)
print('x_pad:\n', x_pad)
print('lengths:', lengths.tolist())
````

## 4) Embeddings: matriz entrenable + lookup
Una capa de embeddings es una matriz:
$$E \in \mathbb{R}^{|\mathcal{V}|\times d}$$
y el embedding de un token `i` es la fila `E[i]`.

En PyTorch: `nn.Embedding(V, d)`.

````python
d_model = 16
emb = nn.Embedding(num_embeddings=len(vocab.itos), embedding_dim=d_model, padding_idx=vocab.pad_id).to(device)

x = x_pad.to(device)  # (B,T)
y = emb(x)            # (B,T,d)
print('x:', x.shape, '-> y:', y.shape)
````

### Lookup explícito (para entender qué hace Embedding)
Esto es equivalente conceptualmente a indexar una matriz.

````python
E = emb.weight  # (V,d)
token_id = x[0,0].item()
print('token:', vocab.itos[token_id], '| id:', token_id)
print('E[token_id] shape:', E[token_id].shape)
print('embedding igual a emb(x)[0,0]? ', torch.allclose(E[token_id], y[0,0]))
````

## 5) Interpretación geométrica: similitud coseno
En embeddings, tokens con contextos similares tienden a acabar cerca.
Aquí solo veremos la mecánica (estos embeddings aún son aleatorios).


````python
def cosine(u, v, eps=1e-8):
    return float((u @ v) / (u.norm()*v.norm() + eps))

w1, w2 = 'pizza', 'NLP'
id1 = vocab.stoi.get(w1, vocab.unk_id)
id2 = vocab.stoi.get(w2, vocab.unk_id)
print(w1, w2, 'cos:', cosine(emb.weight[id1].detach().cpu(), emb.weight[id2].detach().cpu()))
````

## 6.1) BPE (Byte Pair Encoding)
Idea: empezar con símbolos (caracteres) y **fusionar pares frecuentes** iterativamente.
Tokenización final: aplica las fusiones (greedy).

````python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

# Creamos un tokenizador BPE con token para desconocidos
# Si aparece algo no representable por el vocabulario, usará [UNK]
tok_bpe = Tokenizer(BPE(unk_token='[UNK]'))

# Pre-tokenizador por espacios: separa palabras por whitespace antes de aplicar BPE
tok_bpe.pre_tokenizer = Whitespace()

# Entrenador BPE:
# - vocab_size: tamaño máximo del vocabulario final
# - min_frequency: frecuencia mínima para conservar un token
# - special_tokens: tokens reservados con ids fijos
trainer_bpe = BpeTrainer(
    vocab_size=250,
    min_frequency=1,
    special_tokens=['[PAD]', '[UNK]', '[CLS]', '[SEP]', '[MASK]']
)

# Entrenamos BPE usando el corpus de subpalabras que preparamos antes
tok_bpe.train(['sub_corpus.txt'], trainer_bpe)

# Texto de prueba con una palabra larga/rara
text = 'electroencefalografista aprende NLP'

# encode aplica tokenización y nos devuelve ids + tokens
enc = tok_bpe.encode(text)

# Mostramos texto original
print('TEXT:', text)

# Lista de sub-tokens aprendidos para ese texto
print('TOKENS:', enc.tokens)

# decode reconstruye texto aproximado desde ids
print('DECODE:', tok_bpe.decode(enc.ids))
````

## 7) Embeddings posicionales (intuición)
En Transformers, la atención no sabe el orden por sí misma.
Se suma un vector posicional $p_t$ al embedding del token:
$$x_t = e_{i_t} + p_t.$$

Aquí implementamos **positional embeddings aprendibles** (más simple que sinusoidales).

````python
class TinyPositionalEmbedding(nn.Module):
    # max_len: número máximo de posiciones que queremos soportar
    # d_model: dimensión de cada vector posicional (debe coincidir con embedding de token)
    def __init__(self, max_len: int, d_model: int):
        super().__init__()

        # Tabla aprendible de posiciones: fila i = vector de la posición i
        # Forma: (max_len, d_model)
        self.pos = nn.Embedding(max_len, d_model)

    def forward(self, x):
        # x llega como ids de tokens con forma (B, T)
        B, T = x.shape

        # Creamos índices de posición [0, 1, 2, ..., T-1]
        # device=x.device evita errores CPU/GPU mezclados
        positions = torch.arange(T, device=x.device).unsqueeze(0).expand(B, T)

        # Devolvemos vectores posicionales para cada posición del batch
        # Forma de salida: (B, T, d_model)
        return self.pos(positions)

# Máxima longitud que vamos a usar en esta demo
max_len = 32

# Instanciamos el módulo de posiciones y lo movemos al mismo device
pos_emb = TinyPositionalEmbedding(max_len=max_len, d_model=d_model).to(device)

# x_pad contiene ids de tokens con padding (B, T)
x = x_pad.to(device)

# Embedding de tokens: qué palabra es cada posición
tok_vecs = emb(x)

# Embedding posicional: en qué lugar de la secuencia está cada token
pos_vecs = pos_emb(x)

# Entrada final para el modelo = token + posición
x_in = tok_vecs + pos_vecs

# Verificamos shapes para entender el flujo
print('tok_vecs:', tok_vecs.shape)
print('pos_vecs:', pos_vecs.shape)
print('input final:', x_in.shape)
````

## 1) Repaso: tipos de datos → sesgos inductivos
- **Tabular**: vector sin orden → MLP
- **Espacial (2D)**: localidad/traslación → CNN
- **Secuencial**: orden + dependencia temporal → RNN/Transformer

### Por qué MLP no es ideal en PLN
- Longitud fija
- No modela orden
- No comparte parámetros por posición

### Por qué CNN mejora pero sigue limitada
- Capta n-gramas (localidad)
- Paralelizable
- Pero el **contexto efectivo** crece lento (receptive field)
- Dependencias largas son difíciles de capturar de forma natural

## 2) ¿Qué es una RNN?
Una RNN es una **misma célula** aplicada repetidamente:

$$h_t = \tanh(W_x x_t + W_h h_{t-1} + b)$$

- $x_t$: entrada en el tiempo t
- $h_{t-1}$: memoria previa
- $h_t$: nueva memoria

La 'profundidad' de una RNN es **temporal**: el grafo se desenrolla a lo largo de $T$ pasos.

### Many-to-One (clasificación)
$$(x_1,\dots,x_T) \to y$$

### One-to-Many (generación)
$$x \to (y_1,\dots,y_T)$$

### Many-to-Many (etiquetado)
$$(x_1,\dots,x_T) \to (y_1,\dots,y_T)$$

## 3) Implementar una RNN desde cero (célula + desenrollado)
Construimos una célula RNN mínima y la aplicamos en un bucle.

Esto ayuda a ver:
- una sola célula reutilizada
- el estado conecta pasos
- autograd hará BPTT al final

````python
import torch
import torch.nn as nn

class SimpleRNNCell(nn.Module):
    """
    Una celda RNN "básica" (tipo Elman).
    Actualiza el estado oculto h_t usando:
        h_t = tanh(Wx * x_t + Wh * h_{t-1})
    """
    def __init__(self, input_dim: int, hidden_dim: int):
        super().__init__()

        # Proyección de la entrada x_t (D -> H) con bias
        self.Wx = nn.Linear(input_dim, hidden_dim, bias=True)

        # Proyección del estado anterior h_{t-1} (H -> H) sin bias
        # (muchas implementaciones ponen el bias solo en Wx)
        self.Wh = nn.Linear(hidden_dim, hidden_dim, bias=False)

    def forward(self, x_t, h_prev):
        """
        x_t:   (B, D)  entrada en el tiempo t
        h_prev:(B, H)  estado oculto anterior
        devuelve:
        h_t:   (B, H)  nuevo estado oculto
        """
        # Suma de contribución de la entrada y del estado anterior
        preact = self.Wx(x_t) + self.Wh(h_prev)

        # No linealidad típica de RNN simple
        h_t = torch.tanh(preact)
        return h_t


class SimpleRNN(nn.Module):
    """
    RNN que procesa una secuencia completa.
    Recorre T pasos, guarda todos los estados ocultos, y produce logits
    a partir del último estado (many-to-one).
    """
    def __init__(self, input_dim: int, hidden_dim: int, output_dim: int):
        super().__init__()

        # La celda recurrente (actualiza h en cada t)
        self.cell = SimpleRNNCell(input_dim, hidden_dim)

        # Capa final para convertir el último estado oculto en salida/clases
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x, h0=None):
        """
        x:  (B, T, D) batch de secuencias
            B = batch size
            T = longitud temporal
            D = dimensión de entrada (input_dim)

        h0: (B, H) estado inicial opcional
        devuelve:
        logits: (B, output_dim) salida basada en el último estado
        hs:     (B, T, H) todos los estados ocultos a lo largo del tiempo
        """
        B, T, D = x.shape

        # Si no dan h0, inicializamos h a ceros (estado oculto inicial)
        if h0 is None:
            H = self.cell.Wh.in_features  # hidden_dim (entrada de Wh)
            h = torch.zeros(B, H, device=x.device)
        else:
            h = h0

        hs = []  # aquí guardaremos cada h_t

        # Recorremos la secuencia paso a paso
        for t in range(T):
            # Tomamos la entrada del tiempo t: (B, D)
            x_t = x[:, t, :]

            # Actualizamos el estado oculto: (B, H)
            h = self.cell(x_t, h)

            # Guardamos este estado
            hs.append(h)

        # Many-to-one: usamos SOLO el último estado oculto para predecir
        logits = self.fc(hs[-1])  # (B, output_dim)

        # Apilamos la lista hs -> tensor (B, T, H)
        hs = torch.stack(hs, dim=1)

        return logits, hs


print("SimpleRNN ready")
````

## 4) Entrenamiento y BPTT (autograd)
En PyTorch, `loss.backward()` hace BPTT automáticamente.
Inspeccionamos loss y accuracy.

````python
import torch
import torch.nn as nn

def train_parity(model, dl, epochs=10, lr=1e-3, clip=None):
    # Ponemos el modelo en modo entrenamiento:
    # activa dropout/batchnorm (si existieran) y prepara el comportamiento correcto
    model.train()

    # Optimizador: Adam actualiza los parámetros del modelo usando los gradientes
    opt = torch.optim.Adam(model.parameters(), lr=lr)

    # Función de pérdida para clasificación multiclase:
    # espera logits de shape (B, C) y etiquetas y de shape (B,) con valores 0..C-1
    crit = nn.CrossEntropyLoss()

    # Bucle de épocas (pasadas completas por el DataLoader)
    for ep in range(1, epochs + 1):
        total_loss, total_acc, n = 0.0, 0.0, 0  # acumuladores para métricas

        # Iteramos por mini-batches
        for x, y in dl:
            # x: (B, T, 2)  secuencia one-hot
            # y: (B,)       etiqueta (0=par, 1=impar)
            x, y = x.to(device), y.to(device)

            # 1) Resetear gradientes acumulados del paso anterior
            opt.zero_grad()

            # 2) Forward: obtenemos los logits (B, 2)
            #    El modelo devuelve (logits, hs). Aquí hs no lo usamos.
            logits, _ = model(x)

            # 3) Loss: CrossEntropy incluye softmax internamente
            loss = crit(logits, y)

            # 4) Backward: calcula gradientes (BPTT: backprop through time)
            loss.backward()

            # (Opcional) Clipping de gradientes para estabilizar entrenamiento en RNNs
            # Evita explosión de gradientes
            if clip is not None:
                nn.utils.clip_grad_norm_(model.parameters(), clip)

            # 5) Paso del optimizador: actualiza los parámetros con los gradientes
            opt.step()

            # --- Métricas ---
            # loss promedio ponderado por tamaño de batch
            total_loss += loss.item() * y.size(0)

            # accuracy: comparamos clase predicha (argmax) contra y
            preds = logits.argmax(dim=-1)  # (B,)
            total_acc += (preds == y).float().sum().item()

            # contador total de ejemplos vistos
            n += y.size(0)

        # Log cada 2 épocas (y también en la primera)
        if ep % 2 == 0 or ep == 1:
            print(f'ep {ep:02d} | loss {total_loss/n:.4f} | acc {total_acc/n:.3f}')


# Creamos y entrenamos
model = SimpleRNN(input_dim=2, hidden_dim=32, output_dim=2).to(device)
train_parity(model, dl, epochs=10, lr=1e-3, clip=None)
````

Qué significa “BPTT” aquí

Tu RNN guarda la cadena de operaciones para todos los timesteps t=0..T-1.
Cuando haces loss.backward(), PyTorch propaga gradientes desde el último estado hacia atrás por toda la secuencia.

## 5) Vanishing/Exploding gradient: experimento controlado
Simplificación:
$$\frac{\partial h_t}{\partial h_{t-k}} \approx (W_h)^k$$
Si la norma efectiva es <1 → desaparece; >1 → explota.

Aquí definimos: $h_t = \tanh(\alpha h_{t-1})$ y medimos $\|\partial L/\partial h_0\|$.

````python
class ControlledRNNCell(nn.Module):
    """
    Celda RNN artificial MUY simple.
    No tiene pesos aprendibles: solo controla cuánto del estado anterior
    se propaga al siguiente paso mediante el escalar alpha.
    """
    def __init__(self, hidden_dim: int, alpha: float):
        super().__init__()
        self.alpha = alpha          # factor de escala (controla estabilidad)
        self.hidden_dim = hidden_dim

    def forward(self, h_prev):
        # Regla de transición:
        # h_t = tanh(alpha * h_{t-1})
        return torch.tanh(self.alpha * h_prev)

def grad_through_time(alpha: float, T: int = 50, hidden_dim: int = 64):
    # Creamos la celda con un alpha fijo
    cell = ControlledRNNCell(hidden_dim, alpha).to(device)

    # Estado inicial h0, con gradiente activado
    # shape: (1, hidden_dim)
    h0 = torch.randn(1, hidden_dim, device=device, requires_grad=True)

    # Inicializamos el estado
    h = h0

    # Avanzamos T pasos en el tiempo (sin entrada externa)
    for _ in range(T):
        h = cell(h)

    # Definimos una pérdida artificial: suma de todas las componentes finales
    loss = h.sum()

    # Backpropagation Through Time
    loss.backward()

    # Devolvemos la norma del gradiente respecto al estado inicial
    return float(h0.grad.norm().item())

for a in [0.5, 0.9, 1.0, 1.1, 1.5]:
    g = grad_through_time(alpha=a, T=60)
    print(f'alpha={a:3.1f} -> ||dL/dh0|| = {g:.6e}')
````

## 7) Truncated BPTT + detach()
Procesamos la secuencia por chunks y cortamos el grafo para limitar cuánto atrás viaja el gradiente.

````python
def truncated_bptt_last_chunk_grad(T=80, chunk=20, hidden_dim=64, alpha=0.9):
    cell = ControlledRNNCell(hidden_dim, alpha).to(device)

    # Estado inicial (leaf)
    h = torch.randn(1, hidden_dim, device=device, requires_grad=True)

    # Iremos guardando aquí el "inicio del último chunk" como leaf
    h_start_last_leaf = None

    # El último chunk empieza en el paso: T - chunk (0-indexed)
    last_start_step = T - chunk  # ej: T=80, chunk=20 -> empieza en t=60

    for t in range(T):
        # Si estamos justo al inicio del último chunk,
        # hacemos detach para cortar el grafo ANTES de entrar al último chunk
        # y convertimos h en un leaf con requires_grad=True
        if t == last_start_step:
            h = h.detach().requires_grad_(True)
            h_start_last_leaf = h  # este SÍ es leaf

        # Avanzamos un paso
        h = cell(h)

        # Truncamiento normal cada 'chunk' pasos (excepto al final)
        if (t + 1) % chunk == 0 and (t + 1) != T:
            h = h.detach().requires_grad_(True)

    loss = h.sum()
    loss.backward()

    # Si chunk == T, last_start_step = 0, así que h_start_last_leaf se define bien
    return float(h_start_last_leaf.grad.norm().item())


for chunk in [10, 20, 40, 80]:
    g = truncated_bptt_last_chunk_grad(T=80, chunk=chunk, alpha=0.9)
    print(f'chunk={chunk:3d} -> ||dL/dh_start_last|| = {g:.6e}')
````

## 8) LSTM y GRU: clasificación many-to-one
Ahora pasamos al caso PLN típico:
**Embedding → (LSTM/GRU) → último estado → Linear → clase**

Usamos tokenización whitespace para no mezclar temas.

### Fórmulas clave de LSTM y GRU (puente teoría ↔ código)

**LSTM** (idea: memoria explícita con celda $c_t$):
$$
f_t=\sigma(W_f x_t + U_f h_{t-1}+b_f),\quad
 i_t=\sigma(W_i x_t + U_i h_{t-1}+b_i)
$$
$$
\tilde c_t=\tanh(W_c x_t + U_c h_{t-1}+b_c),\quad
c_t=f_t\odot c_{t-1}+i_t\odot \tilde c_t
$$
$$
o_t=\sigma(W_o x_t + U_o h_{t-1}+b_o),\quad
h_t=o_t\odot\tanh(c_t)
$$

**GRU** (menos compuertas, más simple):
$$
z_t=\sigma(W_z x_t + U_z h_{t-1}),\quad
r_t=\sigma(W_r x_t + U_r h_{t-1})
$$
$$
\tilde h_t=\tanh(W_h x_t + U_h(r_t\odot h_{t-1})),\quad
h_t=(1-z_t)\odot h_{t-1}+z_t\odot \tilde h_t
$$

Lectura práctica: ambos mecanismos controlan qué información conservar/olvidar para reducir vanishing gradients en secuencias largas.

````python
from collections import Counter
from dataclasses import dataclass
import torch
from torch.utils.data import Dataset, DataLoader

@dataclass
class Vocab:
    stoi: dict   # string -> id
    itos: list   # id -> string
    pad: str = '<pad>'
    unk: str = '<unk>'
    bos: str = '<bos>'
    eos: str = '<eos>'

    # IDs de tokens especiales
    @property
    def pad_id(self): return self.stoi[self.pad]
    @property
    def unk_id(self): return self.stoi[self.unk]
    @property
    def bos_id(self): return self.stoi[self.bos]
    @property
    def eos_id(self): return self.stoi[self.eos]

def build_vocab(texts, min_freq=1):
    # Cuenta palabras en todos los textos
    c = Counter()
    for t in texts:
        c.update(t.lower().split())

    # Tokens especiales primero (quedan con ids 0..3)
    specials = ['<pad>', '<unk>', '<bos>', '<eos>']

    # Añadimos el resto de palabras que aparecen al menos min_freq veces
    itos = specials + [w for w, f in c.items() if f >= min_freq and w not in specials]
    stoi = {w: i for i, w in enumerate(itos)}

    return Vocab(stoi, itos)

def encode(v: Vocab, text: str, add_bos_eos=True):
    # Tokenización simple por espacios (para demo)
    toks = text.lower().split()

    # Convertimos tokens a ids (si no está, usamos unk_id)
    ids = [v.stoi.get(w, v.unk_id) for w in toks]

    # Añadimos tokens de inicio y fin si se pide
    if add_bos_eos:
        ids = [v.bos_id] + ids + [v.eos_id]
    return ids

def pad_batch(seqs, pad_id):
    # Longitudes reales de cada secuencia
    lens = torch.tensor([len(s) for s in seqs], dtype=torch.long)

    # Longitud máxima en este batch
    T = int(lens.max().item())

    # Matriz (B, T) rellena con pad_id
    out = torch.full((len(seqs), T), pad_id, dtype=torch.long)

    # Copiamos cada secuencia al principio (padding a la derecha)
    for i, s in enumerate(seqs):
        out[i, :len(s)] = torch.tensor(s, dtype=torch.long)

    return out, lens

# Datos de ejemplo
texts = [
    'el perro que persiguio al gato estaba cansado',
    'me gusta la pizza',
    'pytorch entrena redes recurrentes',
    'hoy hace sol en la playa',
    'la rnn mantiene memoria del pasado',
    'cnn capta patrones locales',
    'me encanta aprender nlp',
    'el banco del parque es verde',
    'el banco central sube tipos',
    'las lstm usan compuertas'
]
labels = [1,1,0,1,0,0,0,1,0,0]

# Construimos vocabulario con estos textos
v = build_vocab(texts)

class TextCls(Dataset):
    # Dataset que devuelve (secuencia_ids, label)
    def __init__(self, texts, labels, v):
        self.texts = texts
        self.labels = labels
        self.v = v

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, i):
        # ids = lista de ints (longitud variable)
        ids = encode(self.v, self.texts[i], add_bos_eos=True)
        y = int(self.labels[i])
        return ids, y

def collate(batch):
    # batch es una lista de elementos devueltos por el Dataset
    # cada uno: (lista_ids, y)
    seqs, ys = zip(*batch)

    # Pad a la longitud máxima del batch
    x, l = pad_batch(list(seqs), v.pad_id)

    # Convertimos labels a tensor
    y = torch.tensor(ys, dtype=torch.long)
    return x, l, y

# DataLoader con collate_fn para padding dinámico por batch
dl_text = DataLoader(
    TextCls(texts, labels, v),
    batch_size=4,
    shuffle=True,
    collate_fn=collate
)

# Probamos un batch
xb, lb, yb = next(iter(dl_text))
print('xb:', xb.shape, 'lengths:', lb.tolist(), 'y:', yb.tolist())
````

````python
import torch
import torch.nn as nn

class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, d_model=64, hidden=128, num_classes=2, pad_id=0):
        super().__init__()

        # Embedding: convierte ids (0..vocab_size-1) en vectores de tamaño d_model.
        # padding_idx=pad_id hace que:
        # - la fila del PAD se mantenga fija (no se actualiza)
        # - su gradiente sea 0
        self.emb = nn.Embedding(vocab_size, d_model, padding_idx=pad_id)

        # LSTM (batch_first=True => entrada (B, T, d_model))
        self.lstm = nn.LSTM(d_model, hidden, batch_first=True)

        # Capa final de clasificación
        self.fc = nn.Linear(hidden, num_classes)

    def forward(self, x, lengths):
        """
        x:       (B, T) ids con padding
        lengths: (B,) longitudes reales (sin contar el padding)
        devuelve:
        logits:  (B, num_classes)
        """
        # (B, T) -> (B, T, d_model)
        e = self.emb(x)

        # Empaquetado para ignorar pasos PAD.
        # enforce_sorted=False permite que el batch no esté ordenado por longitud.
        packed = nn.utils.rnn.pack_padded_sequence(
            e, lengths.cpu(), batch_first=True, enforce_sorted=False
        )

        # Pasamos el packed por la LSTM.
        # Para LSTM, la salida final es (hT, cT):
        # hT: (num_layers * num_directions, B, hidden)
        # cT: (num_layers * num_directions, B, hidden)
        _, (hT, cT) = self.lstm(packed)

        # Usamos el hidden final de la última capa: hT[-1] -> (B, hidden)
        # y lo proyectamos a clases: (B, 2)
        logits = self.fc(hT[-1])
        return logits


class GRUClassifier(nn.Module):
    def __init__(self, vocab_size, d_model=64, hidden=128, num_classes=2, pad_id=0):
        super().__init__()

        # Igual: embedding con padding_idx
        self.emb = nn.Embedding(vocab_size, d_model, padding_idx=pad_id)

        # GRU (similar a LSTM pero sin cT)
        self.gru = nn.GRU(d_model, hidden, batch_first=True)

        # Clasificador final
        self.fc = nn.Linear(hidden, num_classes)

    def forward(self, x, lengths):
        # (B, T) -> (B, T, d_model)
        e = self.emb(x)

        # Empaquetado para ignorar padding
        packed = nn.utils.rnn.pack_padded_sequence(
            e, lengths.cpu(), batch_first=True, enforce_sorted=False
        )

        # Para GRU, devuelve:
        # hT: (num_layers * num_directions, B, hidden)
        _, hT = self.gru(packed)

        # Última capa: (B, hidden) -> logits (B, 2)
        logits = self.fc(hT[-1])
        return logits


def train_cls(model, dl, epochs=25, lr=2e-3, clip=1.0):
    # Mandamos el modelo al device (CPU/GPU)
    model.to(device)

    # Optimizador
    opt = torch.optim.Adam(model.parameters(), lr=lr)

    # Loss para clasificación
    crit = nn.CrossEntropyLoss()

    for ep in range(1, epochs + 1):
        model.train()

        tot_loss, tot_acc, n = 0.0, 0.0, 0

        for x, l, y in dl:
            # x: (B, T)  ids padded
            # l: (B,)    longitudes reales
            # y: (B,)    etiquetas
            x, l, y = x.to(device), l.to(device), y.to(device)

            # Reset gradientes
            opt.zero_grad()

            # Forward: logits (B, 2)
            logits = model(x, l)

            # Loss
            loss = crit(logits, y)

            # Backward
            loss.backward()

            # Clipping: evita exploding gradients
            if clip is not None:
                nn.utils.clip_grad_norm_(model.parameters(), clip)

            # Paso del optimizador
            opt.step()

            # Métricas
            tot_loss += loss.item() * y.size(0)
            tot_acc  += (logits.argmax(-1) == y).float().sum().item()
            n += y.size(0)

        # Print cada 5 epochs (y la primera)
        if ep % 5 == 0 or ep == 1:
            print(f'ep {ep:02d} | loss {tot_loss/n:.4f} | acc {tot_acc/n:.3f}')


print('--- LSTM ---')
train_cls(LSTMClassifier(len(v.itos), pad_id=v.pad_id), dl_text)

print('\n--- GRU ---')
train_cls(GRUClassifier(len(v.itos), pad_id=v.pad_id), dl_text)
````

````python
# Etiquetas (según tu ejemplo):
# 1 = "frases tipo cotidiano / narrativa simple"
# 0 = "frases más técnicas (rnn/cnn/lstm/nlp) o 'tech'"
#
# Nota: no existe "verdad universal" aquí; lo importante es que sean consistentes
# con tus labels y que haya más datos para entrenar.

texts_new_pos = [  # clase 1
    "ayer fui al mercado y compre fruta",
    "el gato duerme en el sofa",
    "la comida estaba muy rica",
    "hace frio y llevo abrigo",
    "el parque esta lleno de gente"
]
labels_new_pos = [1]*5

texts_new_neg = [  # clase 0
    "los transformers usan atencion",
    "una gru simplifica la lstm",
    "el embedding representa palabras",
    "backpropagation entrena la red",
    "las convoluciones extraen caracteristicas"
]
labels_new_neg = [0]*5

# Tus datos originales (los que ya tenías)
texts = [
    'el perro que persiguio al gato estaba cansado',
    'me gusta la pizza',
    'pytorch entrena redes recurrentes',
    'hoy hace sol en la playa',
    'la rnn mantiene memoria del pasado',
    'cnn capta patrones locales',
    'me encanta aprender nlp',
    'el banco del parque es verde',
    'el banco central sube tipos',
    'las lstm usan compuertas'
]
labels = [1,1,0,1,0,0,0,1,0,0]

# Unimos todo
texts_aug = texts + texts_new_pos + texts_new_neg
labels_aug = labels + labels_new_pos + labels_new_neg

# Reconstruimos vocab con el dataset ampliado (importante)
v2 = build_vocab(texts_aug, min_freq=1)

# Dataset de texto igual que el tuyo
class TextCls(Dataset):
    def __init__(self, texts, labels, v):
        self.texts = texts
        self.labels = labels
        self.v = v
    def __len__(self):
        return len(self.texts)
    def __getitem__(self, i):
        return encode(self.v, self.texts[i], add_bos_eos=True), int(self.labels[i])

def collate_factory(vocab):
    # Hacemos un collate que "cierra" sobre vocab para no depender de globales
    def collate(batch):
        seqs, ys = zip(*batch)
        x, l = pad_batch(list(seqs), vocab.pad_id)
        y = torch.tensor(ys, dtype=torch.long)
        return x, l, y
    return collate

dl_text_aug = DataLoader(
    TextCls(texts_aug, labels_aug, v2),
    batch_size=4,
    shuffle=True,
    collate_fn=collate_factory(v2)
)

xb, lb, yb = next(iter(dl_text_aug))
print('Aug xb:', xb.shape, 'lengths:', lb.tolist(), 'y:', yb.tolist())


# =========================================================
# Modelos: con y sin pack_padded_sequence
# =========================================================

class LSTMClassifierPacked(nn.Module):
    """
    LSTM many-to-one:
    - Ignora padding usando pack_padded_sequence (recomendado)
    - Clasifica usando h_T (último hidden) de la última capa
    """
    def __init__(self, vocab_size, d_model=64, hidden=128, num_classes=2, pad_id=0):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, d_model, padding_idx=pad_id)
        self.lstm = nn.LSTM(d_model, hidden, batch_first=True)
        self.fc = nn.Linear(hidden, num_classes)

    def forward(self, x, lengths):
        # x: (B,T) ids con PAD
        e = self.emb(x)  # (B,T,d_model)

        # Empaquetamos para que la LSTM NO procese pasos PAD
        packed = nn.utils.rnn.pack_padded_sequence(
            e, lengths.cpu(), batch_first=True, enforce_sorted=False
        )

        # Salida final: hT (1, B, hidden) si num_layers=1 y no bidir
        _, (hT, cT) = self.lstm(packed)

        # hT[-1]: (B, hidden)
        return self.fc(hT[-1])  # (B, num_classes)


class GRUClassifierPacked(nn.Module):
    """
    Igual que LSTM pero con GRU.
    """
    def __init__(self, vocab_size, d_model=64, hidden=128, num_classes=2, pad_id=0):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, d_model, padding_idx=pad_id)
        self.gru = nn.GRU(d_model, hidden, batch_first=True)
        self.fc = nn.Linear(hidden, num_classes)

    def forward(self, x, lengths):
        e = self.emb(x)
        packed = nn.utils.rnn.pack_padded_sequence(
            e, lengths.cpu(), batch_first=True, enforce_sorted=False
        )
        _, hT = self.gru(packed)
        return self.fc(hT[-1])


class LSTMClassifierNoPack(nn.Module):
    """
    Versión SIN pack:
    - La LSTM procesa también los PAD.
    - Si luego usamos el estado del último timestep, ese último timestep puede ser PAD.
    - Aquí usamos output[:, -1, :] como "resumen final" (lo típico si no usas lengths).
    """
    def __init__(self, vocab_size, d_model=64, hidden=128, num_classes=2, pad_id=0):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, d_model, padding_idx=pad_id)
        self.lstm = nn.LSTM(d_model, hidden, batch_first=True)
        self.fc = nn.Linear(hidden, num_classes)

    def forward(self, x, lengths):
        e = self.emb(x)                     # (B,T,d)
        out, (hT, cT) = self.lstm(e)        # out: (B,T,hidden)

        # "Resumen" ingenuo: último timestep del batch (t = T-1)
        # PROBLEMA: para secuencias cortas, ese paso es PAD.
        h_last = out[:, -1, :]              # (B,hidden)
        return self.fc(h_last)


class GRUClassifierNoPack(nn.Module):
    """
    Igual que LSTM NoPack, usando la salida out[:, -1, :].
    """
    def __init__(self, vocab_size, d_model=64, hidden=128, num_classes=2, pad_id=0):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, d_model, padding_idx=pad_id)
        self.gru = nn.GRU(d_model, hidden, batch_first=True)
        self.fc = nn.Linear(hidden, num_classes)

    def forward(self, x, lengths):
        e = self.emb(x)
        out, hT = self.gru(e)
        h_last = out[:, -1, :]
        return self.fc(h_last)


# =========================================================
# Entrenamiento + evaluación simple en el mismo loader
# (para demo; con tan pocos datos, esto mide más "memorización")
# =========================================================

def train_cls(model, dl, epochs=25, lr=2e-3, clip=1.0):
    model.to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    crit = nn.CrossEntropyLoss()

    for ep in range(1, epochs+1):
        model.train()
        tot_loss, tot_acc, n = 0.0, 0.0, 0

        for x, l, y in dl:
            x, l, y = x.to(device), l.to(device), y.to(device)
            opt.zero_grad()
            logits = model(x, l)
            loss = crit(logits, y)
            loss.backward()
            if clip is not None:
                nn.utils.clip_grad_norm_(model.parameters(), clip)
            opt.step()

            tot_loss += loss.item() * y.size(0)
            tot_acc  += (logits.argmax(-1) == y).float().sum().item()
            n += y.size(0)

        if ep % 5 == 0 or ep == 1:
            print(f'ep {ep:02d} | loss {tot_loss/n:.4f} | acc {tot_acc/n:.3f}')


# =========================================================
# 2) Re-entrena y compara LSTM vs GRU (CON pack)
# =========================================================

print('=== CON pack_padded_sequence ===')
print('--- LSTM Packed ---')
train_cls(LSTMClassifierPacked(len(v2.itos), pad_id=v2.pad_id), dl_text_aug, epochs=25, lr=2e-3, clip=1.0)

print('\n--- GRU Packed ---')
train_cls(GRUClassifierPacked(len(v2.itos), pad_id=v2.pad_id), dl_text_aug, epochs=25, lr=2e-3, clip=1.0)


# =========================================================
# 3) Quita pack_padded_sequence y observa cambios
# =========================================================

print('\n=== SIN pack_padded_sequence ===')
print('--- LSTM NoPack ---')
train_cls(LSTMClassifierNoPack(len(v2.itos), pad_id=v2.pad_id), dl_text_aug, epochs=25, lr=2e-3, clip=1.0)

print('\n--- GRU NoPack ---')
train_cls(GRUClassifierNoPack(len(v2.itos), pad_id=v2.pad_id), dl_text_aug, epochs=25, lr=2e-3, clip=1.0)
````

El padding puede sesgar el resumen porque, si no se maneja explícitamente, la RNN sigue actualizando su estado oculto durante los timesteps <pad>.
En un esquema many-to-one que toma el último estado del batch, ese estado puede corresponder a padding y no al final real de la secuencia.
Aunque el embedding del PAD esté congelado, la dinámica recurrente sigue modificando el estado y el modelo puede aprender a usar la cantidad de padding (longitud) como pista espuria.
En datasets pequeños esto puede no afectar la accuracy, pero en problemas reales perjudica la generalización.

## 1) ¿Qué es un modelo de lenguaje?
Un modelo de lenguaje asigna probabilidades a secuencias:
$$P(w_1,\dots,w_T).$$

Intuición: completa huecos *"Hoy hace mucho ___"*.

## 2) Autorregresivo
Un modelo autorregresivo predice la siguiente palabra usando el pasado:
$$P(w_1,\dots,w_T)=\prod_{t=1}^{T}P(w_t\mid w_{<t}).$$

Esto permite **generación**: el modelo se alimenta de sus predicciones.

## 3) Preparar datos: tokenización + IDs
Para centrarnos en el modelo, usaremos tokenización whitespace y un vocabulario pequeño.

Incluimos tokens especiales:
- `<bos>` inicio
- `<eos>` fin
- `<pad>` padding (si hacemos batches de secuencias)
- `<unk>` OOV

En un LM autorregresivo entrenamos con:
- input: `[w1,...,w(T-1)]`
- target: `[w2,...,wT]` (shift)

````python
from collections import Counter
from dataclasses import dataclass
from typing import Dict, List

@dataclass
class Vocab:
    stoi: Dict[str, int]
    itos: List[str]
    pad: str = '<pad>'
    unk: str = '<unk>'
    bos: str = '<bos>'
    eos: str = '<eos>'

    @property
    def pad_id(self): return self.stoi[self.pad]
    @property
    def unk_id(self): return self.stoi[self.unk]
    @property
    def bos_id(self): return self.stoi[self.bos]
    @property
    def eos_id(self): return self.stoi[self.eos]


def build_vocab(texts: List[str], min_freq: int = 1) -> Vocab:
    c = Counter()
    for t in texts:
        c.update(t.lower().split())
    specials = ['<pad>', '<unk>', '<bos>', '<eos>']
    itos = specials + [w for w, f in c.items() if f >= min_freq and w not in specials]
    stoi = {w: i for i, w in enumerate(itos)}
    return Vocab(stoi=stoi, itos=itos)


def encode(v: Vocab, text: str, add_bos_eos: bool = True) -> List[int]:
    toks = text.lower().split()
    ids  = [v.stoi.get(w, v.unk_id) for w in toks]
    if add_bos_eos:
        ids = [v.bos_id] + ids + [v.eos_id]
    return ids


def decode(v: Vocab, ids: List[int], skip_specials: bool = True) -> str:
    specials = {v.pad, v.unk, v.bos, v.eos}
    toks = []
    for i in ids:
        w = v.itos[i] if 0 <= i < len(v.itos) else '<oov>'
        if skip_specials and w in specials:
            continue
        toks.append(w)
    return ' '.join(toks)

````

## 4) Dataset autorregresivo (shift)
Vamos a convertir el corpus en una secuencia larga de IDs y crear pares (x,y) por ventanas.

Esto es típico para LM con RNN: **truncated BPTT** implícito por ventanas.

- `block_size = T` define cuántos tokens de contexto usamos.
- `x = ids[i:i+T]`
- `y = ids[i+1:i+T+1]`

````python
# -------------------------
# 2) Dataset (ventana deslizante)
# -------------------------
class LMDataset(Dataset):
    """
    Dado un stream de ids, genera pares (x, y) por ventanas:
        x = ids[i : i+block]        <- input
        y = ids[i+1 : i+block+1]   <- target (shift de 1)

    Esto implementa teacher forcing: en cada posición t,
    el modelo predice ids[t+1] dado ids[0..t].
    """
    def __init__(self, ids: torch.Tensor, block_size: int):
        self.ids   = ids
        self.block = block_size

    def __len__(self):
        return max(0, len(self.ids) - self.block - 1)

    def __getitem__(self, idx):
        x = self.ids[idx     : idx + self.block]      # (T,)
        y = self.ids[idx + 1 : idx + self.block + 1]  # (T,)  ← shift
        return x, y


# -------------------------
# 3) Modelo RNNLM
# -------------------------
class RNNLM(nn.Module):
    """
    Embedding  →  Dropout  →  GRU/LSTM  →  Dropout  →  Linear(vocab)

    forward(x, h) devuelve (logits, h_nuevo):
        - logits : (B, T, vocab_size)  <- una predicción por posición
        - h_nuevo: estado oculto para continuar la secuencia
    """
    def __init__(
        self,
        vocab_size: int,
        emb_dim:    int   = 32,
        hidden_dim: int   = 64,
        num_layers: int   = 1,
        rnn_type:   str   = 'gru',   # 'gru' | 'lstm'
        dropout:    float = 0.0,
        pad_id:     int   = 0,
    ):
        super().__init__()
        self.rnn_type = rnn_type.lower()

        # Embedding: convierte ids -> vectores densos
        self.emb      = nn.Embedding(vocab_size, emb_dim, padding_idx=pad_id)
        self.drop_in  = nn.Dropout(dropout)

        # Capa recurrente
        rnn_cls = nn.GRU if self.rnn_type == 'gru' else nn.LSTM
        self.rnn = rnn_cls(
            input_size  = emb_dim,
            hidden_size = hidden_dim,
            num_layers  = num_layers,
            batch_first = True,
            dropout     = dropout if num_layers > 1 else 0.0,
        )

        self.drop_out = nn.Dropout(dropout)

        # Proyección al vocabulario
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x: torch.Tensor, h=None):
        # x : (B, T)
        e      = self.drop_in(self.emb(x))   # (B, T, emb_dim)
        out, h = self.rnn(e, h)              # out: (B, T, hidden_dim)
        out    = self.drop_out(out)
        logits = self.fc(out)                # (B, T, vocab_size)
        return logits, h
````

````python
# -------------------------
# 4) Train / Eval
# -------------------------
@torch.no_grad()
def evaluate(model, dl, pad_id) -> Tuple[float, float]:
    model.eval()
    total_loss, total_tok = 0.0, 0
    for x, y in dl:
        x, y   = x.to(device), y.to(device)
        logits, _ = model(x)
        loss = F.cross_entropy(
            logits.reshape(-1, logits.size(-1)),
            y.reshape(-1),
            reduction='sum',
            ignore_index=pad_id,
        )
        total_loss += loss.item()
        total_tok  += (y != pad_id).sum().item()
    nll = total_loss / max(1, total_tok)
    return nll, math.exp(min(20, nll))
````

````python
def train(
    model, train_dl, valid_dl,
    pad_id,
    lr:        float = 1e-2,
    epochs:    int   = 15,
    grad_clip: float = 1.0,
    patience:  int   = 3,
):
    opt  = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-2)
    best_val, no_imp = float('inf'), 0

    for ep in range(1, epochs + 1):
        model.train()
        run_loss, run_tok = 0.0, 0

        for x, y in train_dl:
            x, y = x.to(device), y.to(device)
            logits, _ = model(x)
            loss = F.cross_entropy(
                logits.reshape(-1, logits.size(-1)),
                y.reshape(-1),
                reduction='sum',
                ignore_index=pad_id,
            )
            opt.zero_grad(set_to_none=True)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            opt.step()
            run_loss += loss.item()
            run_tok  += (y != pad_id).sum().item()

        tr_nll = run_loss / max(1, run_tok)
        vl_nll, vl_ppl = evaluate(model, valid_dl, pad_id)

        print(f"epoch {ep:02d} | "
              f"train nll {tr_nll:.3f} ppl {math.exp(min(20,tr_nll)):.1f} | "
              f"valid nll {vl_nll:.3f} ppl {vl_ppl:.1f}")

        # Early stopping
        if vl_nll < best_val:
            best_val = vl_nll
            torch.save(model.state_dict(), 'best_lm.pt')
            no_imp = 0
        else:
            no_imp += 1
            if no_imp >= patience:
                print(f"  → early stop en epoch {ep}")
                model.load_state_dict(torch.load('best_lm.pt'))
                break

    return model
````

````python
# -------------------------
# 5) Generación autoregresiva
# -------------------------
@torch.no_grad()
def generate(
    model, vocab: Vocab,
    prompt: str,
    max_new: int   = 10,
    temperature: float = 0.8,
    top_k: int     = 10,
) -> str:
    model.eval()

    # Warm-up: pasamos el prompt completo para obtener el estado oculto
    ids = [vocab.bos_id] + [vocab.stoi.get(w, vocab.unk_id)
                             for w in prompt.lower().split()]
    x   = torch.tensor(ids, dtype=torch.long, device=device).unsqueeze(0)
    logits, h = model(x)

    last = x[:, -1:]   # último token como arranque
    generated = ids[:]

    for _ in range(max_new):
        logits, h = model(last, h)
        next_logits = logits[:, -1, :] / max(temperature, 1e-6)

        # Top-k filtering
        if top_k > 0:
            v, ix = torch.topk(next_logits, k=min(top_k, next_logits.size(-1)))
            mask = torch.full_like(next_logits, float('-inf'))
            mask.scatter_(1, ix, v)
            next_logits = mask

        probs  = F.softmax(next_logits, dim=-1)
        next_id = torch.multinomial(probs, num_samples=1)

        tid = int(next_id.item())
        generated.append(tid)
        last = next_id
        if tid == vocab.eos_id:
            break

    return decode(vocab, generated, skip_specials=True)
````

## Entrenamiento autorregresivo (teacher forcing)
Entrenamiento típico:
- `x` ya es contexto
- `y` es el target
- Loss: CrossEntropy sobre todos los pasos

Medimos también **perplejidad**:
$$\mathrm{PPL} = \exp(\text{loss promedio})$$
cuando loss es NLL por token.

````python
# --- Vocabulario del corpus ---
# Construimos mapeo token<->id para entrenar el LM
vocab = build_vocab(corpus, min_freq=1)
print(f"Vocab size: {len(vocab.itos)}")

# --- Convertimos corpus a stream continuo de ids ---
# En un LM autoregresivo trabajamos con una secuencia larga de tokens
all_ids = []
for s in corpus:
    all_ids.extend(encode(vocab, s, add_bos_eos=True))
ids_tensor = torch.tensor(all_ids, dtype=torch.long)
print(f"Total tokens: {len(ids_tensor)}")

# --- Mostramos el shift x->y (teacher forcing) ---
# x contiene contexto, y contiene el "siguiente token" en cada posición
ds_demo = LMDataset(ids_tensor, block_size=6)
x0, y0 = ds_demo[0]
print("\nShift x→y (teacher forcing):")
print("  x:", decode(vocab, x0.tolist(), skip_specials=False))
print("  y:", decode(vocab, y0.tolist(), skip_specials=False))

# --- DataLoaders ---
block_size = 8
train_dl = DataLoader(LMDataset(ids_tensor, block_size), batch_size=4, shuffle=True)
valid_dl = DataLoader(LMDataset(ids_tensor, block_size), batch_size=4, shuffle=False)

# --- Modelo RNNLM ---
model = RNNLM(
    vocab_size=len(vocab.itos),
    emb_dim=32,
    hidden_dim=64,
    num_layers=1,
    rnn_type='gru',
    dropout=0.0,
    pad_id=vocab.pad_id,
).to(device)

# --- Forward de prueba para inspeccionar formas ---
x_b, y_b = next(iter(train_dl))
logits, h = model(x_b.to(device))
print(f"\nForward pass — logits: {logits.shape}")  # (B, T, V)

# Pérdida inicial (antes de entrenar) para tener referencia
loss0 = F.cross_entropy(logits.reshape(-1, logits.size(-1)), y_b.to(device).reshape(-1))
print(
    f"Loss inicial: {loss0.item():.3f}  PPL: {math.exp(loss0.item()):.1f}  "
    f"(≈ log({len(vocab.itos)}) = {math.log(len(vocab.itos)):.2f} si aleatorio)"
)

# --- Entrenamiento ---
print("\n--- Entrenamiento ---")
model = train(model, train_dl, valid_dl, pad_id=vocab.pad_id, lr=1e-2, epochs=20, patience=4)

# --- Generación ---
print("\n--- Generación ---")
for prompt in ["el gato", "hoy hace", "las rnn"]:
    out = generate(model, vocab, prompt, max_new=8, temperature=0.7, top_k=5)
    print(f"  '{prompt}' → '{out}'")
````

## Ejercicios y recurso del uso de RNN como LSTM/GRU en PLN: Modelo de lenguaje (siguiente palabra) con GRU → cámbialo a LSTM y compara

### Objetivo
Ahora practicamos un modelo de **lenguaje autorregresivo** (next-token prediction) usando un dataset público real:** WikiText-2.**

- Entrada: `x = (w1, w2, ..., w_{T-1})`
- Target: `y = (w2, w3, ..., w_T)`
- La red produce logits **en cada paso temporal**, y entrenamos con `CrossEntropyLoss` token a token.
- Se evalúa con: **Negative Log-Likelihood (NLL)** y **Perplexity (PPL)**

### Dataset:
Utilizaremos
**WikiText-2 (raw)**: https://huggingface.co/datasets/Salesforce/wikitext (usamores los 2500 primeros textos de train para el entrenamiento y 500 de validación por la limitación de GPU)

### Arquitectura del modelo
Embedding → RNN (GRU o LSTM) → Linear → Softmax

````python
import math
import random
from collections import Counter
from dataclasses import dataclass
from typing import Dict, List, Tuple, Optional

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from datasets import load_dataset

# -------------------------
# 1) Vocab + encode/decode
# -------------------------
@dataclass
class VocabLM:
    stoi: Dict[str, int]
    itos: List[str]
    pad: str = "<pad>"
    unk: str = "<unk>"
    bos: str = "<bos>"
    eos: str = "<eos>"

    @property
    def pad_id(self): return self.stoi[self.pad]
    @property
    def unk_id(self): return self.stoi[self.unk]
    @property
    def bos_id(self): return self.stoi[self.bos]
    @property
    def eos_id(self): return self.stoi[self.eos]


def build_vocab_lm(texts: List[str], min_freq: int = 2, max_vocab: int = 50000) -> VocabLM:
    c = Counter()
    for t in texts:
        c.update(t.lower().split())

    specials = ["<pad>", "<unk>", "<bos>", "<eos>"]
    # ordenamos por frecuencia (desc) y luego lexicográfico para estabilidad
    words = sorted([(w, f) for w, f in c.items() if w not in specials and f >= min_freq],
                   key=lambda x: (-x[1], x[0]))
    words = [w for w, _ in words][: max(0, max_vocab - len(specials))]

    itos = specials + words
    stoi = {w: i for i, w in enumerate(itos)}
    return VocabLM(stoi=stoi, itos=itos)


def encode_lm(v: VocabLM, text: str, add_bos_eos: bool = True) -> List[int]:
    toks = text.lower().split()
    ids = [v.stoi.get(w, v.unk_id) for w in toks]
    if add_bos_eos:
        ids = [v.bos_id] + ids + [v.eos_id]
    return ids


def decode_lm(v: VocabLM, ids: List[int], skip_specials: bool = True) -> str:
    specials = {v.pad, v.unk, v.bos, v.eos}
    toks = []
    for i in ids:
        w = v.itos[i] if 0 <= i < len(v.itos) else v.unk
        if skip_specials and w in specials:
            continue
        toks.append(w)
    return " ".join(toks)
````

````python
# -------------------------
# 2) Dataset LM por ventanas
# -------------------------
class LMDataset(Dataset):
    def __init__(self, ids: torch.Tensor, block_size: int):
        self.ids = ids
        self.block = block_size

    def __len__(self):
        return max(0, len(self.ids) - self.block - 1)

    def __getitem__(self, idx):
        x = self.ids[idx: idx + self.block]           # (T,)
        y = self.ids[idx + 1: idx + self.block + 1]   # (T,)
        return x, y
````

````python
# -------------------------
# 3) Modelo: Embedding + (GRU|LSTM) + Capa lineal
# -------------------------
class RNNLM(nn.Module):
    def __init__(
        self,
        vocab_size: int,
        emb_dim: int = 128,
        hidden_dim: int = 256,
        num_layers: int = 1,
        rnn_type: str = "gru",      # "gru" o "lstm"
        dropout: float = 0.0,
        pad_id: int = 0
    ):
        super().__init__()

        # Guardamos tamaño de vocab para referencia
        self.vocab_size = vocab_size

        # Embedding: convierte cada id entero en un vector denso de tamaño emb_dim
        self.emb = nn.Embedding(vocab_size, emb_dim, padding_idx=pad_id)

        # Dropout de entrada: regulariza los embeddings durante entrenamiento
        self.drop_in = nn.Dropout(dropout)

        # Normalizamos tipo de RNN a minúsculas para evitar errores por mayúsculas
        rnn_type = rnn_type.lower()
        self.rnn_type = rnn_type

        # Elegimos arquitectura recurrente según parámetro
        if rnn_type == "gru":
            self.rnn = nn.GRU(
                input_size=emb_dim,
                hidden_size=hidden_dim,
                num_layers=num_layers,
                batch_first=True,
                dropout=dropout if num_layers > 1 else 0.0
            )
        elif rnn_type == "lstm":
            # LSTM mantiene dos estados internos: h_t (oculto) y c_t (celda)
            # Eso suele ayudar a modelar dependencias más largas
            self.rnn = nn.LSTM(
                input_size=emb_dim,
                hidden_size=hidden_dim,
                num_layers=num_layers,
                batch_first=True,
                dropout=dropout if num_layers > 1 else 0.0
            )
        else:
            raise ValueError("rnn_type debe ser 'gru' o 'lstm'")

        # Proyección final: hidden_dim -> vocab_size (logits por palabra)
        self.fc = nn.Linear(hidden_dim, vocab_size)

        # Dropout de salida recurrente
        self.drop_out = nn.Dropout(dropout)

    def forward(self, x: torch.Tensor, h=None):
        # x tiene forma (B, T): batch de secuencias de ids

        # 1) Lookup de embeddings: (B, T) -> (B, T, emb_dim)
        e = self.drop_in(self.emb(x))

        # 2) Paso por la RNN (GRU o LSTM)
        # out: representaciones por timestep
        # h: estado final (para seguir generando en inferencia)
        out, h = self.rnn(e, h)

        # 3) Dropout en la salida temporal
        out = self.drop_out(out)

        # 4) Capa lineal para obtener logits sobre todo el vocabulario
        logits = self.fc(out)

        return logits, h
````

### Modelo base (GRU) - EJEMPLO


````python
# -------------------------
# 4) Entrenamiento y evaluación (NLL + Perplexity)
# -------------------------
@torch.no_grad()
def evaluate(model: nn.Module, dl: DataLoader, device: str, pad_id: int) -> Tuple[float, float]:
    # Modo evaluación: desactiva dropout y usa comportamiento determinista
    model.eval()

    # Acumuladores globales para pérdida total y número de tokens válidos
    total_loss = 0.0
    total_tokens = 0

    for x, y in dl:
        # x: entradas, y: targets desplazados una posición
        x, y = x.to(device), y.to(device)

        # logits: (B, T, V) donde V = tamaño del vocabulario
        logits, _ = model(x)

        # CrossEntropy por token (reduction='sum' para acumular NLL total)
        # ignore_index evita penalizar posiciones PAD
        loss = F.cross_entropy(
            logits.reshape(-1, logits.size(-1)),
            y.reshape(-1),
            reduction="sum",
            ignore_index=pad_id
        )

        total_loss += loss.item()
        total_tokens += (y != pad_id).sum().item()

    # NLL media por token
    avg_nll = total_loss / max(1, total_tokens)

    # Perplexity = exp(NLL)
    # limitamos valor para evitar overflow numérico en casos extremos
    ppl = math.exp(min(20, avg_nll))
    return avg_nll, ppl


def train_one_model(
    rnn_type: str,
    train_dl: DataLoader,
    valid_dl: DataLoader,
    vocab_size: int,
    pad_id: int,
    device: str,
    emb_dim: int = 128,
    hidden_dim: int = 256,
    num_layers: int = 1,
    dropout: float = 0.0,
    lr: float = 2e-3,
    epochs: int = 5,
    grad_clip: float = 1.0,
    seed: int = 123
):
    # Fijamos semilla para comparar GRU y LSTM de forma justa
    torch.manual_seed(seed)
    random.seed(seed)

    # Construimos el modelo con el tipo de RNN solicitado
    model = RNNLM(
        vocab_size=vocab_size,
        emb_dim=emb_dim,
        hidden_dim=hidden_dim,
        num_layers=num_layers,
        rnn_type=rnn_type,
        dropout=dropout,
        pad_id=pad_id
    ).to(device)

    # AdamW: Adam + regularización por weight decay
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=0.5)

    for ep in range(1, epochs + 1):
        model.train()
        running = 0.0
        tokens = 0

        for x, y in train_dl:
            x, y = x.to(device), y.to(device)

            logits, _ = model(x)

            # Pérdida de lenguaje: siguiente token correcto en cada timestep
            loss = F.cross_entropy(
                logits.reshape(-1, logits.size(-1)),
                y.reshape(-1),
                reduction="sum",
                ignore_index=pad_id
            )

            # Backprop
            opt.zero_grad(set_to_none=True)
            loss.backward()

            # Gradient clipping para evitar exploding gradients
            if grad_clip is not None:
                nn.utils.clip_grad_norm_(model.parameters(), grad_clip)

            opt.step()

            running += loss.item()
            tokens += (y != pad_id).sum().item()

        train_nll = running / max(1, tokens)
        train_ppl = math.exp(min(20, train_nll))

        val_nll, val_ppl = evaluate(model, valid_dl, device, pad_id)

        print(f"[{rnn_type.upper()}] epoch {ep:02d} | "
              f"train_nll {train_nll:.4f} ppl {train_ppl:.2f} | "
              f"valid_nll {val_nll:.4f} ppl {val_ppl:.2f}")

    return model
````

### ¿Qué estamos optimizando en un LM? NLL y Perplexity

Para una secuencia $w_{1:T}$, un modelo autorregresivo factoriza:
$$
P(w_{1:T})=\prod_{t=1}^{T} P(w_t\mid w_{<t})
$$

La pérdida usada en entrenamiento es la **negative log-likelihood media**:
$$
\text{NLL}= -\frac{1}{T}\sum_{t=1}^{T}\log P_\theta(w_t\mid w_{<t})
$$

Y la métrica interpretativa es:
$$
\text{Perplexity}=\exp(\text{NLL})
$$

Intuición: menor perplexity significa que el modelo está "menos sorprendido" por los tokens reales.

````python
# -------------------------
# 5) Inferencia: generación autoregresiva
# -------------------------
@torch.no_grad()
def generate_next_words(
    model: nn.Module,
    vocab: VocabLM,
    prompt: str,
    max_new_tokens: int = 20,
    temperature: float = 1.0,
    top_k: Optional[int] = 50,
    device: str = "cpu"
) -> str:
    """
    Generación token a token:
    1) Codificamos el prompt
    2) Ejecutamos el modelo para obtener estado oculto inicial
    3) Muestreamos siguiente token repetidamente
    """
    model.eval()

    # Prompt -> ids, añadiendo BOS al inicio
    prompt_ids = [vocab.bos_id] + [vocab.stoi.get(w, vocab.unk_id) for w in prompt.lower().split()]

    # Tensor shape (1, T): batch de tamaño 1
    ids = torch.tensor(prompt_ids, dtype=torch.long, device=device).unsqueeze(0)

    # Warm-up: pasamos el prompt completo para obtener contexto oculto
    h = None
    logits, h = model(ids, h)

    # Último token del prompt será la primera entrada de la fase de generación
    last_id = ids[:, -1:]

    # Guardamos todos los ids generados
    generated = prompt_ids[:]

    for _ in range(max_new_tokens):
        # Un paso autoregresivo: entrada = último token
        logits, h = model(last_id, h)

        # Tomamos la distribución del último paso temporal
        next_logits = logits[:, -1, :]

        # Temperature: >1 más aleatorio, <1 más conservador
        next_logits = next_logits / max(1e-6, temperature)

        # Top-k: limita opciones a los k tokens con mayor probabilidad
        if top_k is not None and top_k > 0:
            v, ix = torch.topk(next_logits, k=min(top_k, next_logits.size(-1)))
            mask = torch.full_like(next_logits, float("-inf"))
            mask.scatter_(1, ix, v)
            next_logits = mask

        # Convertimos logits a probabilidades
        probs = F.softmax(next_logits, dim=-1)

        # Muestreamos un token según la distribución
        next_id = torch.multinomial(probs, num_samples=1)

        tid = int(next_id.item())
        generated.append(tid)

        # Realimentamos ese token como entrada del siguiente paso
        last_id = next_id

        # Si aparece EOS, paramos la generación
        if tid == vocab.eos_id:
            break

    # Volvemos de ids a texto
    return decode_lm(vocab, generated, skip_specials=True)
````

````python
def train_and_inference():
    # Elegimos GPU si está disponible para acelerar entrenamiento
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print("Device:", device)

    # Cargamos WikiText-2 raw desde Hugging Face Datasets
    ds = load_dataset("wikitext", "wikitext-2-raw-v1")

    # Filtramos líneas vacías del split de train
    train_texts_all = [t for t in ds["train"]["text"] if t.strip()]

    # Subconjunto para entrenar rápido en entorno educativo
    train_texts = train_texts_all[:1000]

    # Split validación: también quitamos vacíos
    valid_texts_all = [t for t in ds["validation"]["text"] if t.strip()]

    # Usamos una porción de valid para no disparar tiempo de cómputo
    valid_texts = valid_texts_all[100]

    print("Train size:", len(train_texts))
    print("Valid size:", len(valid_texts))

    # Construimos vocabulario desde todo train (mejor cobertura léxica)
    vocab = build_vocab_lm(train_texts_all, min_freq=2, max_vocab=30000)
    print("Vocab size:", len(vocab.itos))

    # Convierte lista de textos en un stream continuo de ids
    def texts_to_stream(texts: List[str]) -> torch.Tensor:
        all_ids = []
        for t in texts:
            all_ids.extend(encode_lm(vocab, t, add_bos_eos=True))
        return torch.tensor(all_ids, dtype=torch.long)

    train_ids = texts_to_stream(train_texts)
    valid_ids = texts_to_stream(valid_texts)
    print("Train tokens:", len(train_ids), "| Valid tokens:", len(valid_ids))

    # Hiperparámetros compartidos para comparación justa GRU vs LSTM
    block_size = 32
    batch_size = 64
    epochs = 1
    lr = 2e-3
    emb_dim = 128
    hidden_dim = 128
    num_layers = 1
    dropout = 0.4
    grad_clip = 1.0

    # DataLoader para lenguaje autoregresivo por ventanas
    train_dl = DataLoader(LMDataset(train_ids, block_size), batch_size=batch_size, shuffle=True, drop_last=True)
    valid_dl = DataLoader(LMDataset(valid_ids, block_size), batch_size=batch_size, shuffle=False)

    # 1) Modelo base: GRU
    print("\n=== Entrenando GRU ===")
    gru_model = train_one_model(
        "gru",
        train_dl, valid_dl,
        vocab_size=len(vocab.itos),
        pad_id=vocab.pad_id,
        device=device,
        emb_dim=emb_dim,
        hidden_dim=hidden_dim,
        num_layers=num_layers,
        dropout=dropout,
        lr=lr,
        epochs=epochs,
        grad_clip=grad_clip,
        seed=123
    )

    # 2) Modelo alternativo: LSTM (mismos hiperparámetros)
    print("\n=== Entrenando LSTM ===")
    lstm_model = train_one_model(
        "lstm",
        train_dl, valid_dl,
        vocab_size=len(vocab.itos),
        pad_id=vocab.pad_id,
        device=device,
        emb_dim=emb_dim,
        hidden_dim=hidden_dim,
        num_layers=num_layers,
        dropout=dropout,
        lr=lr,
        epochs=epochs,
        grad_clip=grad_clip,
        seed=123
    )

    # 3) Evaluación final en validación
    gru_nll, gru_ppl = evaluate(gru_model, valid_dl, device, vocab.pad_id)
    lstm_nll, lstm_ppl = evaluate(lstm_model, valid_dl, device, vocab.pad_id)

    print("\n=== Comparación final (VALID) ===")
    print(f"GRU  : nll={gru_nll:.4f}  ppl={gru_ppl:.2f}")
    print(f"LSTM : nll={lstm_nll:.4f}  ppl={lstm_ppl:.2f}")

    if lstm_ppl < gru_ppl:
        print("→ LSTM obtiene menor perplejidad: modela mejor la distribución.")
    else:
        print("→ GRU obtiene menor perplejidad (o similar): más eficiente para este dataset pequeño.")

    # 4) Generación de texto con ambos modelos
    prompt = "the meaning of"
    print("\n--- Generación (GRU) ---")
    print(generate_next_words(gru_model, vocab, prompt, max_new_tokens=25, temperature=0.9, top_k=50, device=device))

    print("\n--- Generación (LSTM) ---")
    print(generate_next_words(lstm_model, vocab, prompt, max_new_tokens=25, temperature=0.9, top_k=50, device=device))

    # Guardamos checkpoints para reutilizarlos luego
    torch.save({"model": gru_model.state_dict(), "vocab": vocab}, "gru_lm.pt")
    torch.save({"model": lstm_model.state_dict(), "vocab": vocab}, "best_lm.pt")
    print("\nGuardado: gru_lm.pt y best_lm.pt (LSTM)")
````

# Recurso adicional: Usar RNN para series temporales

En este bloque trabajaremos un problema de predicción de series temporales utilizando una red LSTM. El objetivo es entrenar un modelo que, dado un conjunto de valores pasados de una serie, sea capaz de predecir el siguiente valor. Para ello utilizaremos un dataset público real: Daily Minimum Temperatures in Melbourne (1981–1990), que contiene temperaturas mínimas diarias registradas durante 10 años.

````python
# Operaciones matemáticas (por ejemplo, raíz cuadrada para RMSE)
import math

# pandas para leer y manipular datos tabulares (CSV)
import pandas as pd

# PyTorch para modelo, tensores y entrenamiento
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# -------------------------
# 1) Cargar dataset público (CSV)
# -------------------------
# Dataset: temperaturas mínimas diarias de Melbourne
# Si no quieres depender de internet, puedes descargar el CSV y usar ruta local
CSV_URL = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"

# Leemos el CSV en un DataFrame de pandas
df = pd.read_csv(CSV_URL)

# Nos quedamos con la columna de temperatura como float32 para PyTorch
series = df["Temp"].astype("float32").values
print("Total puntos:", len(series))

# -------------------------
# 2) Split train/valid/test (time-based)
# -------------------------
# n = número total de observaciones temporales
n = len(series)

# Dividimos por tiempo (nunca mezclamos futuro con pasado)
train_end = int(n * 0.7)
valid_end = int(n * 0.85)

# Particiones cronológicas
train_raw = series[:train_end]
valid_raw = series[train_end:valid_end]
test_raw  = series[valid_end:]

# Normalización usando SOLO train (evita fuga de información)
mu = train_raw.mean()
sigma = train_raw.std() + 1e-8

# norm: pasa a escala estandarizada (media ~0, std ~1)
def norm(x):
    return (x - mu) / sigma

# denorm: vuelve de escala normalizada a grados reales
def denorm(x):
    return x * sigma + mu

# Aplicamos normalización a cada split
train = norm(train_raw)
valid = norm(valid_raw)
test  = norm(test_raw)

# -------------------------
# 3) Dataset de ventanas
# -------------------------
class WindowDataset(Dataset):
    """
    Convierte una serie 1D en pares (x, y):
    - x: ventana de longitud T
    - y: siguiente valor inmediato
    """
    def __init__(self, arr, window=30):
        # Guardamos la serie como tensor float32
        self.arr = torch.tensor(arr, dtype=torch.float32)
        self.window = window

    def __len__(self):
        # Cada ejemplo consume "window" puntos y predice el siguiente
        return max(0, len(self.arr) - self.window)

    def __getitem__(self, idx):
        # x: secuencia histórica de tamaño window
        x = self.arr[idx: idx + self.window]          # (T,)

        # y: valor siguiente a esa ventana
        y = self.arr[idx + self.window]               # escalar

        # Añadimos dimensión de feature para LSTM: (T,) -> (T,1)
        # y también dejamos y con forma (1,)
        return x.unsqueeze(-1), y.unsqueeze(-1)

# Hiperparámetros de ventana y batch
window = 30
batch_size = 64

# DataLoaders
train_dl = DataLoader(WindowDataset(train, window), batch_size=batch_size, shuffle=True)
valid_dl = DataLoader(WindowDataset(valid, window), batch_size=batch_size, shuffle=False)
test_dl  = DataLoader(WindowDataset(test, window),  batch_size=batch_size, shuffle=False)

# -------------------------
# 4) Modelo LSTM para regresión 1-step
# -------------------------
class LSTMForecaster(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=64, num_layers=1, dropout=0.0):
        super().__init__()

        # LSTM: procesa secuencias y mantiene memoria temporal
        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )

        # Capa lineal para mapear estado oculto -> predicción escalar
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x, h=None):
        # x: (B, T, 1)

        # out: estados en todos los timesteps
        # h: estado final (h_n, c_n)
        out, h = self.lstm(x, h)      # out: (B, T, H)

        # Tomamos el último timestep como resumen de la ventana
        last = out[:, -1, :]          # (B, H)

        # Predicción final del siguiente valor
        yhat = self.fc(last)          # (B, 1)
        return yhat, h

# Selección de dispositivo
device = "cuda" if torch.cuda.is_available() else "cpu"

# Instanciamos el modelo
model = LSTMForecaster(hidden_dim=64, num_layers=1).to(device)

# -------------------------
# 5) Train/Eval
# -------------------------
# AdamW para optimizar pesos
opt = torch.optim.AdamW(model.parameters(), lr=1e-3)

# MSE para regresión
loss_fn = nn.MSELoss()

@torch.no_grad()
def eval_rmse(dl):
    # Modo evaluación
    model.eval()

    # Suma de errores cuadrados y contador de muestras
    se_sum = 0.0
    n = 0

    for x, y in dl:
        x, y = x.to(device), y.to(device)
        yhat, _ = model(x)

        # Error cuadrático acumulado
        se_sum += ((yhat - y) ** 2).sum().item()
        n += y.numel()

    # RMSE = sqrt(error cuadrático medio)
    rmse = math.sqrt(se_sum / max(1, n))
    return rmse

# Entrenaremos pocas épocas para demo rápida
epochs = 10
grad_clip = 1.0

for ep in range(1, epochs + 1):
    model.train()
    total = 0.0
    count = 0

    for x, y in train_dl:
        x, y = x.to(device), y.to(device)
        yhat, _ = model(x)
        loss = loss_fn(yhat, y)

        opt.zero_grad(set_to_none=True)
        loss.backward()

        # Clipping para estabilizar gradientes
        nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
        opt.step()

        total += loss.item() * y.numel()
        count += y.numel()

    train_mse = total / max(1, count)
    valid_rmse = eval_rmse(valid_dl)
    print(f"epoch {ep:02d} | train_mse {train_mse:.4f} | valid_rmse {valid_rmse:.4f}")

# Evaluación final en test
test_rmse = eval_rmse(test_dl)
print("\nTEST RMSE (normalizado):", test_rmse)

# -------------------------
# 6) Inferencia: predecir próximos k pasos (autoregresivo)
# -------------------------
@torch.no_grad()
def forecast_k_steps(model, seed_window, k=7):
    """
    Recibe una ventana inicial (normalizada) y predice k pasos hacia delante.
    Cada predicción se realimenta como entrada del siguiente paso.
    """
    model.eval()

    # Convertimos seed_window a tensor con forma (1, T, 1)
    window_vals = torch.tensor(seed_window, dtype=torch.float32, device=device).view(1, -1, 1)

    preds = []
    h = None

    for _ in range(k):
        # Predicción del próximo punto
        yhat, h = model(window_vals, h)          # (1,1)
        next_val = yhat.item()
        preds.append(next_val)

        # Desplazamos ventana: quitamos primer valor, añadimos predicción al final
        new_seq = torch.cat([window_vals[:, 1:, :], yhat.view(1, 1, 1)], dim=1)
        window_vals = new_seq

    return preds

# Ejemplo: arrancamos desde la primera ventana del test
seed = test[:window]
preds_norm = forecast_k_steps(model, seed, k=7)

# Volvemos a escala real (°C)
preds = [denorm(p) for p in preds_norm]
print("\nPredicción próximos 7 días (°C aprox):", [round(p, 2) for p in preds])
````
